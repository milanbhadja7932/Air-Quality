```
import numpy as np
import math
import cv2
import random
import os


def read_img(img_path, grayscale):
    if grayscale:
        im = cv2.imread(img_path, 0)
    else:
        im = cv2.imread(img_path)
    return im


def random_crop(image, new_size):
    h, w = image.shape[:2]
    y = np.random.randint(0, h - new_size)
    x = np.random.randint(0, w - new_size)
    image = image[y:y+new_size, x:x+new_size]
    return image


def rotate_image(img, angle, crop):
    h, w = img.shape[:2]
    angle %= 360
    M_rotate = cv2.getRotationMatrix2D((w/2, h/2), angle, 1)
    img_rotated = cv2.warpAffine(img, M_rotate, (w, h))
    if crop:
        angle_crop = angle % 180
        if angle_crop > 90:
            angle_crop = 180 - angle_crop
        theta = angle_crop * np.pi / 180.0
        hw_ratio = float(h) / float(w)
        tan_theta = np.tan(theta)
        numerator = np.cos(theta) + np.sin(theta) * tan_theta
        r = hw_ratio if h > w else 1 / hw_ratio
        denominator = r * tan_theta + 1
        crop_mult = numerator / denominator
        w_crop = int(round(crop_mult*w))
        h_crop = int(round(crop_mult*h))
        x0 = int((w-w_crop)/2)
        y0 = int((h-h_crop)/2)
        img_rotated = img_rotated[y0:y0+h_crop, x0:x0+w_crop]
    return img_rotated


def random_rotate(img, angle_vari, p_crop):
    angle = np.random.uniform(-angle_vari, angle_vari)
    crop = False if np.random.random() > p_crop else True
    return rotate_image(img, angle, crop)


def generate_image_list(args):
    filenames = os.listdir(train_data_dir)
    num_imgs = len(filenames)
    num_ave_aug = int(math.floor(augment_num/num_imgs))
    rem = augment_num - num_ave_aug*num_imgs
    lucky_seq = [True]*rem + [False]*(num_imgs-rem)
    random.shuffle(lucky_seq)

    img_list = [
        (os.sep.join([train_data_dir, filename]), num_ave_aug+1 if lucky else num_ave_aug)
        for filename, lucky in zip(filenames, lucky_seq)
    ]

    return img_list


def augment_images(filelist, args):
    for filepath, n in filelist:
        img = read_img(filepath, grayscale)
        if img.shape[:2] != (im_resize, im_resize):
            img = cv2.resize(img, (im_resize, im_resize))
        filename = filepath.split(os.sep)[-1]
        dot_pos = filename.rfind('.')
        imgname = filename[:dot_pos]
        ext = filename[dot_pos:]

        print('Augmenting {} ...'.format(filename))
        for i in range(n):
            img_varied = img.copy()
            varied_imgname = '{}_{:0>3d}_'.format(imgname, i)
            
            if random.random() < p_rotate:
                img_varied_ = random_rotate(
                    img_varied,
                    rotate_angle_vari,
                    p_rotate_crop)
                if img_varied_.shape[0] >= patch_size and img_varied_.shape[1] >= patch_size:
                    img_varied = img_varied_
                varied_imgname += 'r'

            if random.random() < p_crop:
                img_varied = random_crop(
                    img_varied,
                    patch_size)
                varied_imgname += 'c'

            if random.random() < p_horizonal_flip:
                img_varied = cv2.flip(img_varied, 1)
                varied_imgname += 'h'

            if random.random() < p_vertical_flip:
                img_varied = cv2.flip(img_varied, 0)
                varied_imgname += 'v'

            output_filepath = os.sep.join([
                aug_dir,
                '{}{}'.format(varied_imgname, ext)])
            print("output",output_filepath)
            cv2.imwrite(output_filepath, img_varied)


def get_patch(image, new_size, stride):
    h, w = image.shape[:2]
    i, j = new_size, new_size
    patch = []
    while i <= h:
        while j <= w:
            patch.append(image[i - new_size:i, j - new_size:j])
            j += stride
        j = new_size
        i += stride
    return np.array(patch)


def patch2img(patches, im_size, patch_size, stride):
    img = np.zeros((im_size, im_size, patches.shape[3]+1))
    i, j = patch_size, patch_size
    k = 0
    while i <= im_size:
        while j <= im_size:
            img[i - patch_size:i, j - patch_size:j, :-1] += patches[k]
            img[i - patch_size:i, j - patch_size:j, -1] += np.ones((patch_size, patch_size))
            k += 1
            j += stride
        j = patch_size
        i += stride
    mask=np.repeat(img[:,:,-1][...,np.newaxis], patches.shape[3], 2)
    img = img[:,:,:-1]/mask
    return img


def set_img_color(img, predict_mask, weight_foreground, grayscale):
    if grayscale:
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    origin = img
    img[np.where(predict_mask == 255)] = (0,0,255)
    cv2.addWeighted(img, weight_foreground, origin, (1 - weight_foreground), 0, img)
    return img


def bg_maskc(img, value, mode, grayscale):

    if not grayscale:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _,thresh=cv2.threshold(img,value,255,mode)

    def FillHole(mask):
        contours, hierarchy = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        len_contour = len(contours)
        contour_list = []
        for i in range(len_contour):
            drawing = np.zeros_like(mask, np.uint8)  # create a black image
            img_contour = cv2.drawContours(drawing, contours, i, (255, 255, 255), -1)
            contour_list.append(img_contour)

        out = sum(contour_list)
        return out

    thresh = FillHole(thresh)
    if type(thresh) is int:
        return np.ones(img.shape)
    mask_ = np.ones(thresh.shape)
    mask_[np.where(thresh <= 127)] = 0
    return mask_
```

```
from tensorflow.keras.layers import Input, Conv2D, Conv2DTranspose, LeakyReLU
from tensorflow.keras.models import Model


def AutoEncoder(cfg):
    input_img = Input(shape=(patch_size, patch_size, input_channel))

    h = Conv2D(flc, (4, 4), strides=2, activation=LeakyReLU(alpha=0.2), padding='same')(input_img)
    h = Conv2D(flc, (4, 4), strides=2, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    if patch_size==256:
        h = Conv2D(flc, (4, 4), strides=2, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2D(flc, (3, 3), strides=1, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2D(flc*2, (4, 4), strides=2, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2D(flc*2, (3, 3), strides=1, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2D(flc*4, (4, 4), strides=2, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2D(flc*2, (3, 3), strides=1, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2D(flc, (3, 3), strides=1, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    encoded = Conv2D(z_dim, (8, 8), strides=1, activation='linear', padding='valid')(h)

    h = Conv2DTranspose(flc, (8, 8), strides=1, activation=LeakyReLU(alpha=0.2), padding='valid')(encoded)
    h = Conv2D(flc*2, (3, 3), strides=1, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2D(flc*4, (3, 3), strides=1, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2DTranspose(flc*2, (4, 4), strides=2, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2D(flc*2, (3, 3), strides=1, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2DTranspose(flc, (4, 4), strides=2, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2D(flc, (3, 3), strides=1, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    h = Conv2DTranspose(flc, (4, 4), strides=2, activation=LeakyReLU(alpha=0.2), padding='same')(h)
    if patch_size==256:
        h = Conv2DTranspose(flc, (4, 4), strides=2, activation=LeakyReLU(alpha=0.2), padding='same')(h)

    decoded = Conv2DTranspose(input_channel, (4, 4), strides=2, activation='sigmoid', padding='same')(h)

    return Model(input_img, decoded)
```

```
#aug_dir='./train_patches/bottle'
aug_dir='/content/Autoencoder/Augmented'
augment_num=10000
batch_size=128
bg_mask=None
#chechpoint_dir='./results/bottle/chechpoints/ssim_loss'
chechpoint_dir='/content/Autoencoder/chechpoints/ssim_loss'
decay=1e-05
do_aug=True
epochs=20
flc=32
grayscale=False
im_resize=266
input_channel=3
l1_threshold=None
loss='ssim_loss'
lr=0.0002
mask_size=256
name='bottle'
p_crop=1 
p_horizonal_flip=0.3
p_rotate=0.0
p_rotate_crop=1.0
p_vertical_flip=0.3
patch_size=256
percent=98.0
rotate_angle_vari=45.0
#save_dir='./results/bottle/reconst/ssim_l1_metric_ssim_loss'
save_dir='/content/Autoencoder/reconst/ssim_l1_metric_ssim_loss'
ssim_threshold=None 
stride=32
sub_folder=[]
test_dir='/content/Autoencoder/Untitled Folder/bottle/test'
train_data_dir='/content/Autoencoder/Untitled Folder/bottle/train/good'
weight=1
weight_file=None
z_dim=500
```

```
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.utils import Sequence
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
import numpy as np
import cv2
from glob import glob
import os

#from network import AutoEncoder
#from utils import generate_image_list, augment_images, read_img
#from options import Options


#cfg = Options().parse()

class data_flow(Sequence):
    def __init__(self, filenames, batch_size, grayscale):
        self.filenames = filenames
        self.batch_size = batch_size
        self.grayscale = grayscale

    def __len__(self):
        return int(np.ceil(len(self.filenames) / float(self.batch_size)))

    def __getitem__(self, idx):
        batch_x = self.filenames[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_x = np.array([read_img(filename, self.grayscale) for filename in batch_x])
        
        batch_x = batch_x / 255.
        return batch_x, batch_x

# data
cfg="cfg"
if aug_dir and do_aug:
    print("image",aug_dir)
    img_list = generate_image_list(cfg)
    print(img_list)
    augment_images(img_list, cfg)

dataset_dir = aug_dir if aug_dir else train_data_dir
file_list = glob(dataset_dir + '/*')
num_valid_data = int(np.ceil(len(file_list) * 0.2))
data_train = data_flow(file_list[:-num_valid_data], batch_size, grayscale)
data_valid = data_flow(file_list[-num_valid_data:], batch_size, grayscale)

# loss
if loss == 'ssim_loss':
    
    @tf.function
    def ssim_loss(gt, y_pred, max_val=1.0):
        return 1 - tf.reduce_mean(tf.image.ssim(gt, y_pred, max_val=max_val))
    
    loss = ssim_loss
elif loss == 'ssim_l1_loss':
    
    @tf.function
    def ssim_l1_loss(gt, y_pred, max_val=1.0):
        ssim_loss = 1 - tf.reduce_mean(tf.image.ssim(gt, y_pred, max_val=max_val))
        L1 = tf.reduce_mean(tf.abs(gt - y_pred))
        return ssim_loss + L1 * weight
    
    loss = ssim_l1_loss
else:
    loss = 'mse'

# network
autoencoder = AutoEncoder(cfg)
optimizer = Adam(lr=lr, decay=decay)
autoencoder.compile(optimizer=optimizer, loss=loss, metrics=['mae'] if loss == 'mse' else ['mse'])
autoencoder.summary()

earlystopping = EarlyStopping(patience=20)

checkpoint = ModelCheckpoint(os.path.join(chechpoint_dir, '{epoch:02d}-{val_loss:.5f}.hdf5'), save_best_only=True,
                            period=1, mode='auto', verbose=1, save_weights_only=True)

print("data valid",data_valid)

autoencoder.fit(data_train, epochs=20, validation_data=data_valid, callbacks=[checkpoint, earlystopping])

# show reconstructed images
decoded_imgs = autoencoder.predict(data_valid)
n = len(decoded_imgs)
save_snapshot_dir = chechpoint_dir +'/snapshot/'
if not os.path.exists(save_snapshot_dir):
    os.makedirs(save_snapshot_dir)
for i in range(n):
    cv2.imwrite(save_snapshot_dir+str(i)+'_rec_valid.png', (decoded_imgs[i]*255).astype('uint8'))
```

```
import numpy as np
from skimage.metrics import structural_similarity as ssim
from skimage import morphology 
from glob import glob
import cv2
import os

#from utils import read_img, get_patch, patch2img, set_img_color, bg_mask
#from network import AutoEncoder
#from options import Options


#cfg = Options().parse()
cfg="cfg"

# network
autoencoder = AutoEncoder(cfg)

if weight_file:
    autoencoder.load_weights(chechpoint_dir + '/' + weight_file)
else:
    file_list = os.listdir(chechpoint_dir)
    latest_epoch = max([int(i.split('-')[0]) for i in file_list if 'hdf5' in i])
    print('load latest weight file: ', latest_epoch)
    autoencoder.load_weights(glob(chechpoint_dir + '/' + str(latest_epoch) + '*.hdf5')[0])
#autoencoder.summary()

def get_residual_map(img_path, cfg):
    print("get resudal map")
    test_img = read_img(img_path, grayscale)

    if test_img.shape[:2] != (im_resize, im_resize):
        test_img = cv2.resize(test_img, (im_resize, im_resize))
    if im_resize != mask_size:
        tmp = (im_resize - mask_size)//2
        test_img = test_img[tmp:tmp+mask_size, tmp:tmp+mask_size]

    test_img_ = test_img / 255.

    if test_img.shape[:2] == (patch_size, patch_size):
        test_img_ = np.expand_dims(test_img_, 0)
        decoded_img = autoencoder.predict(test_img_)
    else:
        patches = get_patch(test_img_, patch_size, stride)
        patches = autoencoder.predict(patches)
        decoded_img = patch2img(patches, im_resize, patch_size, stride)

    rec_img = np.reshape((decoded_img * 255.).astype('uint8'), test_img.shape)

    if grayscale:
        ssim_residual_map = 1 - ssim(test_img, rec_img, win_size=11, full=True)[1]
        l1_residual_map = np.abs(test_img / 255. - rec_img / 255.)
    else:
        ssim_residual_map = ssim(test_img, rec_img, win_size=11, full=True, multichannel=True)[1]
        ssim_residual_map = 1 - np.mean(ssim_residual_map, axis=2)
        l1_residual_map = np.mean(np.abs(test_img / 255. - rec_img / 255.), axis=2)

    return test_img, rec_img, ssim_residual_map, l1_residual_map


def get_threshold(cfg):
    print('estimating threshold...')
    global ssim_threshold
    global l1_threshold
    valid_good_list = glob(train_data_dir + '/*png')
    num_valid_data = int(np.ceil(len(valid_good_list) * 0.2))
    total_rec_ssim, total_rec_l1 = [], []
    for img_path in valid_good_list[-num_valid_data:]:
        _, _, ssim_residual_map, l1_residual_map = get_residual_map(img_path, cfg)
        total_rec_ssim.append(ssim_residual_map)
        total_rec_l1.append(l1_residual_map)
    total_rec_ssim = np.array(total_rec_ssim)
    total_rec_l1 = np.array(total_rec_l1)
    ssim_threshold = float(np.percentile(total_rec_ssim, [percent]))
    l1_threshold = float(np.percentile(total_rec_l1, [percent]))
    print('ssim_threshold: %f, l1_threshold: %f' %(ssim_threshold, l1_threshold))
    if not ssim_threshold:
        ssim_threshold = ssim_threshold
    if not l1_threshold:
        l1_threshold = l1_threshold


def get_depressing_mask(cfg):
    global cfdepr_mask
    print("get depressing mask")
    depr_mask = np.ones((mask_size, mask_size)) * 0.2
    depr_mask[5:mask_size-5, 5:mask_size-5] = 1
    cfdepr_mask = depr_mask
    print("cfd",cfdepr_mask)


def get_results(file_list, cfg):
    #print("get result")
    #print("cfdefer",cfdepr_mask)
    #print(file_list)
    for img_path in file_list:
        #print("inside for")
        img_name = img_path.split('\\')[-1][:-4]
        c = '' if not sub_folder else k
        test_img, rec_img, ssim_residual_map, l1_residual_map = get_residual_map(img_path, cfg)

        ssim_residual_map *= cfdepr_mask
        #print("loss",loss)
        if 'ssim' in ls:
            print("inside if")
            l1_residual_map *= cfdepr_mask

        mask = np.zeros((mask_size, mask_size))
        #print("ssim",ssim_residual_map)
        #print("ssim thresold",ssim_threshold)
        mask[ssim_residual_map > ssim_threshold] = 1
        mask[l1_residual_map > l1_threshold] = 1
        if bg_mask == 'B':
            bg_m = bg_maskc(test_img.copy(), 50, cv2.THRESH_BINARY, grayscale)
            mask *= bg_m
        elif bg_mask == 'W':
            #print("type",(test_img.copy()))
            #print("type2",(grayscale))
            #print("type3",(cv2.THRESH_BINARY_INV))
            bg_m = bg_maskc(test_img.copy(), 200, cv2.THRESH_BINARY_INV, grayscale)
            #print("bg",bg_m)
            mask *= bg_m
        kernel = morphology.disk(4)
        mask = morphology.opening(mask, kernel)
        mask *= 255

        vis_img = set_img_color(test_img.copy(), mask, weight_foreground=0.3, grayscale=grayscale)
        save_output="/content/Autoencoder/output"
        cv2.imwrite(save_output+'_residual.png', mask)
        print("mask")
        cv2.imwrite(save_output+'_origin.png', test_img)
        cv2.imwrite(save_output+'_rec.png', rec_img)
        cv2.imwrite(save_output+'_visual.png', vis_img)


        cv2.imwrite(save_dir+'/'+c+'_'+img_name+'_residual.png', mask)
        cv2.imwrite(save_dir+'/'+c+'_'+img_name+'_origin.png', test_img)
        cv2.imwrite(save_dir+'/'+c+'_'+img_name+'_rec.png', rec_img)
        cv2.imwrite(save_dir+'/'+c+'_'+img_name+'_visual.png', vis_img)



```
